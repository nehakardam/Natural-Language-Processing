# Part A: Sentiment lexicon-based classifier.

Sentiment lexicon-based classifier. Create a classifier using a sentiment lexicon. A lexicon from Hu and Liu [1] is provided in the directory opinion lexicon English, but you are welcome to find and use (with attribution, of course) another. Tokenize the data, and classify each document as positive if and only if it has more positive sentiment words than negative sentiment words. Compute and report the accuracy and F1 score (on detecting positive reviews) on the test set, using this lexicon-based classifier.

```python
def result_of_sentiment_analysis(review):
    pos_count = 0
    neg_count = 0
    for token in review:
        if token in dict_neg_pos:
            s = dict_neg_pos.get(token)
            if s == "positive":
                pos_count = pos_count + 1
            else:
                neg_count = neg_count + 1
    if pos_count >= neg_count:
        return "Positive"
    else: 
        return "Negative"
    
sentiments_y = []
reviews_tokens = Testset_data['Reviews_Tokenized']
for review in reviews_tokens:
    sentiments_y.append(result_of_sentiment_analysis(review))
```

```python
Testset_data['Sentiments_Y'] = sentiments_y
Testset_data
```

## Accuracy and F1 Score calculation

```python
actual = Testset_data['Sentiments']
predicted = Testset_data['Sentiments_Y']

tp = np.sum((actual=='Positive') & (predicted=='Positive'))
fp = np.sum((actual!='Positive') & (predicted=='Positive'))
tn = np.sum((actual=='Negative') & (predicted=='Negative'))
fn = np.sum((actual!='Negative') & (predicted=='Negative'))

accuracy = (tp+tn)/(tp+tn+fp+fn)
precision = tp/(tp+fp)
recall = tp/(tp+fn)
f1 = 2 * (precision * recall) / (precision + recall)
```

```python
print (accuracy, precision, recall, f1)
```

## Bag of word Model

```python
word_dict = {}

def add_all_workds_to_dict(words):
    for word in words:
        if word in word_dict:
            word_dict[word] = word_dict.get(word) + 1
        else:
            word_dict[word] = 1

for words in Training_data["Reviews_Tokenized"]:
    add_all_workds_to_dict(words)
```

```python
frequent_word_set = set()
for key, value in word_dict.items():
    if value > 10:
        frequent_word_set.add(key)
```

```python
Training_data_with_features = pd.DataFrame(columns = frequent_word_set)
for review in Training_data["Reviews_Tokenized"]:
    i = len(Training_data_with_features)
    Training_data_with_features.loc[i] = 0
    for word in review:
        if word in frequent_word_set:
            count = Training_data_with_features.loc[i, word]
            Training_data_with_features.loc[i, word] = count + 1
train_X = Training_data_with_features.astype(float) 
```

```python
train_Y = []
for sentiment in Training_data["Sentiments"]:
    if sentiment == "Positive":
        train_Y.append(1)
    else:
        train_Y.append(0)
```

## Logistic Regression Model Classifier 

```python
class Logistic_Regression() : 
    def __init__(self, learning_rate=0.001, n_iters=1000):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        for i in range(self.n_iters):
            model_linear = np.dot(X, self.weights) + self.bias
            predict_y = self._sigmoid(model_linear)
            dw = (1 / n_samples) * np.dot(X.T, (predict_y - y)) 
            db = (1 / n_samples) * np.sum(predict_y - y) 
            self.weights -= self.lr * dw
            self.bias -= self.lr * db

    def predict(self, X):
        model_linear = np.dot(X, self.weights) + self.bias
        predict_y = self._sigmoid(model_linear)
        predict_y_cls = [1 if i > 0.5 else 0 for i in predict_y]
        return np.array(predict_y_cls)

    def _sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
```

```python
regression = Logistic_Regression()
regression.fit(train_X, train_Y)
```

```python
Testset_data_with_features = pd.DataFrame(columns = frequent_word_set)
for review in Testset_data["Reviews_Tokenized"]:
    i = len(Testset_data_with_features)
    Testset_data_with_features.loc[i] = 0
    for word in review:
        if word in frequent_word_set:
            count = Testset_data_with_features.loc[i, word]
            Testset_data_with_features.loc[i, word] = count + 1
test_X = Testset_data_with_features.astype(float) 
```

```python
test_Y = []
for sentiment in Testset_data["Sentiments"]:
    if sentiment == "Positive":
        test_Y.append(1)
    else:
        test_Y.append(0)
```
